```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Data Description
The ZHVI is Zillowâ€™s monthly time series tracking typical home values across different states in the USA, beginning in January 2000. It aims to provide a more accurate view of home value trends than traditional indices like median sale price or repeat sales indices, which are prone to bias from changing sales composition.

```python
data=pd.read_csv("ZHVI.csv")
```

```python
data.info()
data.describe()
```

```python
data.head()
```

# Data Cleaning

```python
data.isnull().sum()
```

```python
data = data.bfill().ffill()
data.isnull().sum()
```

```python
data.rename(columns={data.columns[0]: 'Month'}, inplace=True)
```

```python
# Convert the 'Month' column to datetime objects
data['Month'] = pd.to_datetime(data['Month'])
```

```python
data.set_index('Month', inplace=True)
```

```python
data.columns = data.columns.str.lower().str.replace(" ", "_")

string_columns = list(data.dtypes[data.dtypes == "object"].index)
for column in string_columns:
    data[column] = data[column].str.lower().str.replace(" ", "_")

states = data.dtypes[data.dtypes == "float"].index

for state in states:
    data[state] = round(data[state], 2)
```

# Data Representation

```python
from itertools import cycle
colors = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])  # this is a collection of 10 colors that repeats until you need them to

fig, axs = plt.subplots(17, 3, figsize=(15, 30), sharex=True, sharey=False)
axs = axs.flatten()

for i in range(len(states)):
    data[states[i]].plot(title=states[i], color=next(colors), ax=axs[i])

plt.tight_layout()
plt.show()
```

We notice an overall incrasing trend for all the states.

# Insights

## Historical % Growth for each state

```python
# Calculate total growth percentage for each state from first to last valid index
growth_summary = {}

for state in states:
    start = data[state].loc[data[state].first_valid_index()]
    end = data[state].loc[data[state].last_valid_index()]
    growth = ((end - start) / start) * 100
    growth_summary[state] = round(growth, 2)

# Convert to DataFrame
growth_df = pd.DataFrame.from_dict(growth_summary, orient='index', columns=['% Growth Since 2000'])
growth_df = growth_df.sort_values(by='% Growth Since 2000', ascending=False)

# Display top/bottom performers
print("Top 5 Growing States Since 2000:")
print(growth_df.head())

print("\nBottom 5 Growing States Since 2000:")
print(growth_df.tail())
```

```python
# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(data=growth_df.head(5).reset_index(), x='index', y='% Growth Since 2000', palette='Greens_d')
plt.title('Top 5 States with Highest Home Value Growth Since 2000')
plt.xlabel('State')
plt.ylabel('% Growth')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=growth_df.tail(5).reset_index(), x='index', y='% Growth Since 2000', palette='Reds_d')
plt.title('Bottom 5 States with Lowest Home Value Growth Since 2000')
plt.xlabel('State')
plt.ylabel('% Growth')
plt.show()
```

## Percentage Change in the last  years

```python
recent_growth_summary = {}
cutoff_date = data.index.max() - pd.DateOffset(years=5)

for state in states:
    try:
        start = data[state].loc[data[state].index >= cutoff_date].iloc[0]
        end = data[state].iloc[-1]
        recent_growth = ((end - start) / start) * 100
        recent_growth_summary[state] = round(recent_growth, 2)
    except:
        continue  # skip if data is missing

recent_df = pd.DataFrame.from_dict(recent_growth_summary, orient='index', columns=['% Growth Last 5 Years'])
recent_df = recent_df.sort_values(by='% Growth Last 5 Years', ascending=False)

# Show top 5
print("Top 5 States with Fastest Growth in Last 5 Years:")
print(recent_df.head())
```

# Forecasting

```python
from sklearn.linear_model import LinearRegression

# Set forecast period (60 months = 5 years)
months_ahead = 60

# Create a dictionary to store results
forecast_results = {}

# Loop over each state
for state in states:
    y = data[state].dropna()
    X = np.arange(len(y)).reshape(-1, 1)

    # Train model
    model = LinearRegression()
    model.fit(X, y)

    # Predict future
    future_X = np.arange(len(y), len(y) + months_ahead).reshape(-1, 1)
    future_y = model.predict(future_X)

    # Store forecasts with date index
    future_dates = pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=months_ahead, freq='MS')
    forecast_series = pd.Series(future_y, index=future_dates)

    forecast_results[state] = forecast_series
```

```python
# Combine historical + forecasted for a sample of states
sample_states = growth_df.head(6).index.tolist()  # top 6 growing states

plt.figure(figsize=(15, 10))
for state in sample_states:
    full_series = pd.concat([data[state], forecast_results[state]])
    plt.plot(full_series.index, full_series.values, label=state)

plt.title("ZHVI Forecast (Next 5 Years) for Top Growing States")
plt.xlabel("Year")
plt.ylabel("ZHVI")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
forecast_growth = {}
for state in states:
    current_val = data[state].iloc[-1]
    future_val = forecast_results[state].iloc[-1]
    growth = ((future_val - current_val) / current_val) * 100
    forecast_growth[state] = round(growth, 2)

forecast_growth_df = pd.DataFrame.from_dict(forecast_growth, orient='index', columns=["Forecasted Growth (5 Years)"])
forecast_growth_df = forecast_growth_df.sort_values(by="Forecasted Growth (5 Years)", ascending=False)

# Display top and bottom
print("Top 5 States (Forecasted Growth Next 5 Years):")
print(forecast_growth_df.head())

print("\nBottom 5 States (Forecasted Growth Next 5 Years):")
print(forecast_growth_df.tail())
```

```python
# Top 5
plt.figure(figsize=(10, 6))
sns.barplot(data=forecast_growth_df.head(5).reset_index(), x='index', y="Forecasted Growth (5 Years)", palette='Blues_d')
plt.title("Top 5 States Forecasted to Grow Most (Next 5 Years)")
plt.xlabel("State")
plt.ylabel("% Growth")
plt.show()

# Bottom 5
plt.figure(figsize=(10, 6))
sns.barplot(data=forecast_growth_df.tail(5).reset_index(), x='index', y="Forecasted Growth (5 Years)", palette='Oranges_d')
plt.title("Bottom 5 States Forecasted to Grow Least (Next 5 Years)")
plt.xlabel("State")
plt.ylabel("% Growth")
plt.show()
```

# K-Means Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Use the last 10 years for clustering
years = 10
subset = data.tail(12 * years).copy()

# Drop states with any missing values in this period
subset = subset.dropna(axis=1)

# Transpose to shape (states, months)
X = subset.T.values

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method to find optimal k
inertia = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
# Using the elbow method, I was able to conclude that the optimal k is 4
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Map states to their cluster
cluster_map = pd.DataFrame({'State': subset.columns, 'Cluster': clusters})
print(cluster_map.sort_values("Cluster"))
```

```python
# PCA for 2D visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='Set2')
for i, state in enumerate(subset.columns):
    plt.text(X_pca[i, 0], X_pca[i, 1], state, fontsize=8)
plt.title("State Clusters Based on ZHVI Trends")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```
